```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("e-commercebasedataset.csv")
df.head()
```

```python
df=df.drop('SALES_MODEL',axis=1)
df=df.drop('PART_NUMBER',axis=1)
df=df.drop('INVOICE_DATE',axis=1)
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
df=df.drop_duplicates()
```

```python
df['Interaction']=1

utility_matrix =df.pivot_table(index='INVOICE_ID', columns='PART_NAME', values='Interaction', fill_value=0)
```

```python
utility_matrix.head()
```

```python
trans=utility_matrix.T
trans.head()
```

```python
from sklearn.decomposition import TruncatedSVD
SVD = TruncatedSVD(n_components=10)
decomposed_matrix = SVD.fit_transform(trans)
decomposed_matrix.shape
decomposed_matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
d_matrix=pd.DataFrame(decomposed_matrix)

cosine_sim_matrix = cosine_similarity(decomposed_matrix)
#print(cosine_sim_matrix)

similarity_matrix = pd.DataFrame(cosine_sim_matrix, index=utility_matrix.columns, columns=utility_matrix.columns)


similarity_matrix.head()
```

```python
ls=list()
ls.append("Accumulators, Receivers, and Dryers")
ls.append("Angles and Channels")
ls.append("Alarm")
ls.append("Wheels")
```

```python
recommendations = []
for product in ls:
    correlation_product_ID = similarity_matrix[product]
    #print(correlation_product_ID)
    recommend = [x for x in trans.index if correlation_product_ID[x] > 0.60]
    recommendations.append(recommend)
print(recommendations) 
```

```python
n=len(recommendations)
fin_set=set()
for i in range(n):
  ls1=recommendations.pop(0)
  common_recommendations = set(ls1).intersection(*recommendations[1:])
  m=len(common_recommendations)
  if(m>3):
    cnt=0
    for i in common_recommendations:
      fin_set.add(i)
      cnt+=1
      if(cnt==3):
        break
  else:
    for i in common_recommendations:
      fin_set.add(i)
  recommendations.append(ls1)
common_recommendations = list(fin_set)
print(fin_set)
common_recommendations = [product for product in common_recommendations if product not in ls]
N = 20
top_recommendations = common_recommendations[:N]
```
